## Fixing Columns

### FIxing Weight Column

```python
df["Weight"].unique()
```

```python
weights = (df["Weight"].str.endswith("lbs"))
df["Weight"] = df["Weight"].str.strip("kglbs")
df["Weight"] = pd.to_numeric(df["Weight"])
df["Weight"] = df["Weight"].astype(float)
```

```python
df_lbs = df[weights]
df.loc[weights, "Weight"] = round(df_lbs["Weight"] * 0.45359237,1)
```

```python
df
```

```python
df.rename(columns={"Weight":"Weight_kg"},inplace=True)
```

### Fixing Height Column

```python
df.Height.unique()
```

```python
def convert_height_cm(x):
    if "cm" in x:
        return int(x.replace("cm", ""))
    else:
        height = x.split("'")
        return round(int(height[0])*30.48+int(height[1][:-1])*2.54)
```

```python
df["Height"] = df["Height"].apply(convert_height_cm)
```

```python
df.rename(columns={"Height":"Height_cm"},inplace=True)
```

### Fixing Hits Column

```python
df["Hits"].unique()
```

```python
df["Hits"] = df["Hits"].str.strip("K")
```

```python
df['Hits'].fillna(value=0, inplace=True)
```

```python
def clean_hits(hits):
    if float(hits) % 1 !=0:
        return float(hits) * 1000
    else:
        return float(hits)
```

```python
df["Hits"] = df["Hits"].apply(clean_hits)
```

```python
df["Hits"] = df["Hits"].astype(float)
```

```python
df["Hits"] = df["Hits"].astype(int)
```

```python
df
```

### Fixing Value Column

```python
df["Value"] = df["Value"].str.replace("€"," ") 
```

```python
def clean_value(value):
    if value[-1] == "M":
        return float(value[1:-1]) * 1000000
    elif value[-1] == "K":
        return float(value[1:-1]) * 1000
    else:
        return value
```

```python
df["Value"] = df["Value"].apply(clean_value)

df["Value"] = df["Value"].astype(int)
```

### Cleaning Wage Column

```python
df["Wage"] = df["Wage"].str.replace("€"," ") 
```

```python
df["Wage"] = df["Wage"].apply(clean_value)
```

```python
df["Wage"] = df["Wage"].astype(int)
```

### Fixing Release Clause Column

```python
df["Release Clause"] = df["Release Clause"].str.replace("€"," ") 
```

```python
df["Release Clause"] = df["Release Clause"].apply(clean_value)
```

## Removing stars from columns

```python
df['W/F'] = df['W/F'].str.strip("★").astype(int)
df['SM'] = df['SM'].str.strip("★").astype(int)
df['IR'] = df['IR'].str.strip("★").astype(int)
```

## Formating Join Date Column

```python
df['Joined'] = pd.to_datetime(df['Joined'], format='%b %d, %Y')
```

```python
df
```

## Splitting Contract Column

```python
df[['Contract Started','Contract Ended']] = df[df['Contract'].str.contains('~')]['Contract'].str.split('~',expand=True).astype('Int64')
```
